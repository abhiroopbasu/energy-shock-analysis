```python
#  Energy Shock Analysis: COVID and LPG Crisis in India (2020–2026)

#  Objective
# To analyze the impact of COVID-19 and LPG supply disruptions on India's inflation and GDP trends.

#  Context
# India experienced a major economic shock during COVID-19 in 2020, followed by recovery. More recently, geopolitical disruptions affected LPG supply, increasing energy prices and economic pressure.

#  Methodology (Pseudocode)

# 1. Load inflation and GDP datasets  
# 2. Clean and select relevant years  
# 3. Filter data for India  
# 4. Analyze trends over time  
# 5. Compare inflation and GDP patterns  
# 6. Interpret impact of COVID and LPG crisis  
```

```python
#  Data Loading and Cleaning (Pseudocode)

# 1. Import required libraries  
# 2. Load inflation and GDP datasets  
# 3. Remove unnecessary metadata rows  
# 4. Select relevant columns (years)  
# 5. Rename columns for consistency  
# 6. Filter dataset for India  

import pandas as pd
import matplotlib.pyplot as plt

inflation = pd.read_csv('inflation.csv', skiprows=4)
gdp = pd.read_csv('gdp_growth.csv', skiprows=4)

inflation = inflation[['Country Name','2000','2005','2010','2015','2020','2021','2022','2023']]
gdp = gdp[['Country Name','2000','2005','2010','2015','2020','2021','2022','2023']]

inflation.columns = ['Country','2000','2005','2010','2015','2020','2021','2022','2023']
gdp.columns = ['Country','2000','2005','2010','2015','2020','2021','2022','2023']

inflation = inflation[inflation['Country']=='India']
gdp = gdp[gdp['Country']=='India']

#  Inflation Trend Analysis (Pseudocode)

# 1. Define relevant years  
# 2. Extract inflation values for India  
# 3. Plot inflation over time  
# 4. Label axes and title  

years = ['2000','2005','2010','2015','2020','2021','2022','2023']

plt.plot(years, inflation.iloc[0][years])
plt.title('Inflation Trend in India')
plt.xlabel('Year')
plt.ylabel('Inflation (%)')
plt.show()


#  GDP Growth Analysis (Pseudocode)

# 1. Use same years for consistency  
# 2. Extract GDP growth values for India  
# 3. Plot GDP growth over time  
# 4. Label axes and title  


#  Comparative Analysis (Pseudocode)

# 1. Plot inflation and GDP on same graph  
# 2. Add legend for comparison  
# 3. Mark COVID year (2020)  
# 4. Interpret divergence between indicators  


plt.plot(years, gdp.iloc[0][years])
plt.title('GDP Growth in India')
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')
plt.show()

plt.plot(years, inflation.iloc[0][years], label='Inflation')
plt.plot(years, gdp.iloc[0][years], label='GDP Growth')

plt.axvline(x='2020', linestyle='--', label='COVID Shock')

plt.legend()
plt.title('Impact of COVID and LPG Crisis on Economy')
plt.xlabel('Year')
plt.show()
```

```python
#  Insights

# - COVID-19 caused a sharp decline in GDP growth in 2020, reflecting economic contraction  
# - Post-COVID recovery is visible in GDP trends, though uneven  
# - Inflation increased in recent years, particularly during global supply disruptions  
# - LPG supply disruptions contributed to rising energy costs and inflationary pressure  
# - The economy shows resilience but remains vulnerable to external shocks  
```
